```python
# Import the qiskit library
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import EstimatorV2 as Estimator


#量子ビットと古典ビットをそれぞれ1つずつ用意
circuit = QuantumCircuit(1,1)

#量子ビット0の測定結果を古典ビット0に記録
circuit.measure(0,0) 

#Imprt The math library
import math

# 初期状態 |0⟩ から、適切な回転を適用して斜め上の状態を作る
theta = math.pi / 3.5 # 緯度角
phi = math.pi / 5 # 経度角

circuit.ry(theta, 0)  # Y軸回転で緯度 θ を設定
circuit.rz(phi, 0)    # Z軸回転で経度 φ を設定

circuit.draw("mpl")
```

```python
# Import the qiskit library
from qiskit_aer import StatevectorSimulator

#状態ベクトルシミュレータを準備
backend = StatevectorSimulator()

#状態ベクトルシミュレータを準備
result = backend.run(circuit).result().get_statevector(circuit, decimals=3)

# Import the qiskit library
#量子回路の可視化（ブロッホ球表示）
from qiskit.visualization import plot_bloch_multivector

#ブロッホ球の表示
plot_bloch_multivector(result)
```
